```python
import requests
import time
import pandas as pd

# FastAPI endpoint
# endpoint = "http://127.0.0.1:8000/docs#/ask" 
endpoint = "http://127.0.0.1:8000/ask"  # Update if needed

input_file = "data-test.xlsx"
output_file = "generated_data_haiku.xlsx"
df1 = pd.read_excel(input_file, engine="openpyxl")

# Ensure the 'user_input' column exists
if 'user_input' not in df1.columns:
    raise ValueError("The Excel file must contain a column named 'user_input'.")

# Prepare response and timing columns
# df["openai_response"] = ""
# df["claude_response_fast"] = ""
# df["claude_response_slow"] = ""
# df["openai"] = 0.0
# df["claude-fast"] = 0.0
# df["claude-slow"] = 0.0

# Function to get response and time
def get_response_and_time(question):
    payload = {
        "user_id": 4,
        "chat_id": 193,
        "conversation_history": [
            {"role": "user", "content": "hi"},
            {"id": 592, "role": "bot", "content": "Hi there! How can I help you today?", "is_liked": False},
            {"role": "user", "content": question}
        ],
        "question": question
    }

    start_time = time.time()
    try:
        response = requests.post(endpoint, json=payload, verify=False)
        response.raise_for_status()
        answer = response.json().get("answer", "No answer returned")
    except Exception as e:
        answer = f"Error: {e}"
    response_time = round(time.time() - start_time, 3)
    return answer, response_time
```

```python
#openai
# Process each question
answers = []
times = []

for index, row in df1.iterrows():
    question = row["user_input"]
    answer, response_time = get_response_and_time(question)
    answers.append(answer)
    times.append(response_time)
    
# Add results to DataFrame
df1['haiku'] = answers
df1['haiku-time'] = times
```
